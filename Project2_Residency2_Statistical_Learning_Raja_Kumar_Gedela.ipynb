### 2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)

```python
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
```

```python
titan_df[['Old Scheme (in thousands)']].head()
```

```python
titan_df[['New Scheme (in thousands)']].head()
```

```python
t_statistic, p_value = ttest_ind(titan_df[['Old Scheme (in thousands)']], titan_df[['New Scheme (in thousands)']])
print(t_statistic, p_value)
```

#### Null Hypothesis : There is no significant difference between data sets - Old Scheme and New Scheme

#### Alternate Hypothesis : There is a significant difference

```python
# p_value < 0.05 (alpha) => alternative hypothesis:
print ("two-sample t-test p-value=", p_value)
```

##### p_value is 0.49063516 which is greater than 0.05 (alpha), so don't reject the Null Hypothesis. i.e. there is no significant difference between old scheme and new scheme.
